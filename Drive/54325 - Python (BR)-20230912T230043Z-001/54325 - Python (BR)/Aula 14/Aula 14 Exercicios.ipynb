1. Utilizando a base "realtor-data.csv" gerar uma base com os seguintes requisitos: 
 - Eliminar as linhas com missing values em variáveis categóricas. 
 - Substituir missing values das colunas bed e bath por zero. 
 - Substituir missing values das colunas house_size pela mediana. 




```python
import pandas as pd
import numpy as np
```

```python
realtor = pd.read_csv("realtor-data.csv")
realtor.head(3)
```

```python
realtor.info()
```

```python
realtor2 = realtor.dropna(subset = ['status','city',"state","prev_sold_date"])
```

```python
realtor3 = realtor2.fillna({"bed":0,
                            "bath":0,
                            "house_size":realtor2["house_size"].median()})
realtor3 = realtor3.reset_index(drop=True)
realtor3.head(3)
```

```python
realtor3.info()
```

2. Da base anterior, criar a seguinte tabela pivot:

```python
realtor3['prev_sold_date'] = pd.to_datetime(realtor3['prev_sold_date'])
realtor3['prev_sold_year'] = realtor3['prev_sold_date'].dt.year
realtor3.head(3)
```

```python
realtor3.pivot_table(index ='state', 
                    columns='prev_sold_year', 
                    values ='price',
                    aggfunc = 'sum')
```
