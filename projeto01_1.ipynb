```python
from plyer import notification
from datetime import datetime
import sqlite3
```

```python
data = str(datetime.now())
def alerta(status_code,base):
    if status_code >= 500:
        mensagem = 'Erro no servidor'
    elif status_code >= 400:
        mensagem = 'Erro na requisição'
    elif status_code == 200:
        mensagem = 'Requisição concluída com sucesso'
    else:
        print ('Outros tipos de erro')
    notification.notify(
    title=f'ATENÇÃO: {mensagem}'
    , message = str('Carregamento da base %s. \n %s' % (base,data))
    , app_name=''
    , timeout=10)
```

```python
status_code = requisicao2.status_code
base = 'estado sp'
alerta(status_code,base)  
```

```python
requisicao3 = requests.get('https://brasilapi.com.br/api/ibge/municipios/v1/{code}?providers=dados-abertos-br,gov,wikipedia')
status_code = requisicao3.status_code
base = 'municipios'
alerta(status_code,base)
```

Revisões pós correção + Tratamento de dados

```python
def extracao(url,base):
    data = str(datetime.now())
    req = requests.get(url)
    status_code = req.status_code
    data_json = req.json()
    if status_code == 200:
        base = pd.DataFrame(data_json)
        return(base)
    elif status_code >= 400:
        message = 'Erro na requisição'
    elif status_code == 200:
        message = 'Requisição concluída com sucesso'
    else:
        print ('Outros tipos de erro')
    notification.notify(
    title=f'ATENÇÃO: {message}'
    , message = str('Carregamento da base %s. \n %s' % (base,data))
    , app_name=''
    , timeout=10)
```

```python
extracao('https://brasilapi.com.br/api/pix/v1/participants','base_pix')
```

```python
extracao('https://brasilapi.com.br/api/ibge/uf/v1','base_uf')
```

```python
extracao('https://brasilapi.com.br/api/ibge/municipios/v1/sp?providers=dados-abertos-br,gov,wikipedia','base_municipios')
```

```python
extracao('https://brasilapi.com.br/api/ibge/uf/v1/sp','base_estado_sp')
```
