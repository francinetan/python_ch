```python
import pandas as pd
```

```python
df = pd.read_csv('realtor-data.csv')
```

```python
df.info()
```

```python
df.dropna(subset=['status','city','state','prev_sold_date'])
```

```python
var_cat = list(df.columns[df.dtypes == 'object'].values)
```

```python
df.dropna(subset =var_cat)
```

```python
df['bed'] = df['bed'].fillna(0)
df['bath'] = df['bath'].fillna(0)
```

```python
df = df.fillna({"house_size":df["house_size"].median()})
df
```

```python
df.info()
```

```python
df["prev_sold_date"] = pd.to_datetime(df["prev_sold_date"])
```

```python
df.dtypes
```

```python
df['prev_ano_venda'] = df['prev_sold_date'].dt.year
```

```python
df.head()
```

```python
df.pivot_table(index='state', columns='prev_ano_venda', values='price', aggfunc='sum')
```
